```python
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import os
```

```python
file_names_to_plot = ['convolutional_autoencoder.pkl','denoising_autoencoder.pkl','vae(1)-Copy2.pkl']
os.makedirs('result', exist_ok=True)
for f_name in file_names_to_plot:
    with open(f_name,'rb') as f:
        data = pickle.load(f)
    plt.figure()
    plt.title(f_name)
    plt.plot(data['valid_cost'],label='Validation Cost',marker='o')
    plt.plot(data['train_cost'],label='Training Cost',marker='o')
    plt.grid()
    plt.legend()
    plt.savefig('result/single_{0}.png'.format(f_name))
```

```python
import numpy as np
#Comperative plot one for comapring training on different methods second on the validation set
file_names_to_plot = ['convolutional_autoencoder.pkl','denoising_autoencoder.pkl']
plot_names = ['valid','train','everything']
plt.figure('valid')
plt.figure('train')
plt.figure('everything')
os.makedirs('result', exist_ok=True)
for f_name in file_names_to_plot:
    with open(f_name,'rb') as f:
        data = pickle.load(f)
    #The scale could be different. need normalization
    
    data['train_cost'] = np.array( data['train_cost'])
    data['train_cost'] /=data['train_cost'].max()
    
    
    data['valid_cost'] = np.array( data['valid_cost'])
    data['valid_cost'] /=data['valid_cost'].max()
    
    plt.figure('train')
    plt.plot(data['train_cost'],label='{0}_train'.format(f_name),marker='o')
    plt.figure('valid')
    plt.plot(data['valid_cost'],label='{0}_valid'.format(f_name),marker='o')
    plt.figure('everything')
    plt.plot(data['train_cost'],label='{0}_train'.format(f_name),marker='o')
    plt.plot(data['valid_cost'],label='{0}_valid'.format(f_name),marker='o')
for p_name in plot_names:
    plt.figure(p_name)
    plt.grid()
    plt.legend(loc='best')
    plt.title(p_name)
    plt.savefig('result/{0}.png'.format(p_name))
```
